````python

## 📝 Section 1: Environment Setup (15 minutes)

### 1.1 Verify Prerequisites

```bash
# Check K3s cluster
kubectl cluster-info
kubectl get nodes

# Verify application pods
kubectl get pods -n logistics-app

# Check Instana agent
kubectl get pods -n instana-agent
kubectl logs -n instana-agent -l app.kubernetes.io/name=instana-agent --tail=50
```

**Expected Results**:
- All application pods: `Running` with `1/1` ready
- Instana agent pod: `Running` with `1/1` ready
- Agent logs show: "Connected to backend"

### 1.2 Set Environment Variables

```bash
# Export your configuration
export INSTANA_TENANT_URL="https://your-tenant.instana.io"
export INSTANA_API_TOKEN="your-api-token"
export VSI_PUBLIC_IP="your-vsi-ip"
export FRONTEND_URL="http://${VSI_PUBLIC_IP}:30080"
export AUTH_URL="http://${VSI_PUBLIC_IP}:30081"
export ORDER_URL="http://${VSI_PUBLIC_IP}:30082"
```

### 1.3 Test Application Connectivity

```bash
# Test frontend
curl -I $FRONTEND_URL

# Test auth service health
curl $AUTH_URL/health

# Test order service health
curl $ORDER_URL/health
```

---

## 🎨 Section 2: Create Application Perspective (20 minutes)

### 2.1 Access Instana UI

1. Open browser: `https://your-tenant.instana.io`
2. Log in with your credentials
3. Navigate to **Infrastructure** view

### 2.2 Verify Service Discovery

**Steps**:
1. Click **Infrastructure** → **Kubernetes** → **Namespaces**
2. Find `logistics-app` namespace
3. Verify all services are discovered:
   - auth-service
   - order-service
   - shipment-service
   - tracking-service
   - frontend
   - postgres

**Screenshot**: Save as `screenshots/01-discovered-services.png`

### 2.3 Create Application Perspective

**Steps**:
1. Click **Applications** in left menu
2. Click **+ Add** button (top right)
3. Select **Create Application Perspective**
4. Configure:
   - **Name**: `Logistics Application`
   - **Scope**: Click **Add Filter**
   - **Filter Type**: `Kubernetes Namespace`
   - **Operator**: `equals`
   - **Value**: `logistics-app`
5. Click **Create**

**Screenshot**: Save as `screenshots/02-app-perspective-created.png`

### 2.4 Verify Application Perspective

**Steps**:
1. Click on **Logistics Application** in the list
2. Verify you see:
   - Service list (6 services)
   - Dependency map
   - Initial metrics (may be empty if no traffic yet)

**Screenshot**: Save as `screenshots/03-app-perspective-overview.png`
````

## 📈 Section 4: Analyze Microservices Performance (20 minutes)

### 4.1 View Service Metrics

**Steps**:
1. Go to **Applications** → **Logistics Application**
2. Observe populated metrics:
   - **Call Rate**: Requests per second
   - **Mean Latency**: Average response time
   - **Error Rate**: Percentage of failures
3. Click on each service to see detailed metrics

**Key Metrics to Record**:
- Auth Service latency: _____ ms
- Order Service latency: _____ ms
- Shipment Service latency: _____ ms

**Screenshot**: Save as `images/04-service-metrics.png`

### 4.2 Analyze Distributed Traces

**Steps**:
1. Click on **Calls** tab in Application Perspective
2. You should see list of traced requests
3. Click on any trace to see detailed view
4. Observe:
   - Service call sequence
   - Time spent in each service
   - Database queries
   - External API calls

**What to Look For**:
- Total trace duration
- Number of spans
- Service dependencies
- Slowest operations

**Screenshot**: Save as `images/05-distributed-trace.png`

### 4.3 Identify Performance Bottlenecks

**Steps**:
1. Click on **Analytics** tab
2. Select **Slowest Calls** from dropdown
3. Identify services with highest latency
4. Click on a slow call to see trace details

**Analysis Questions**:
- Which service has highest average latency?
- Are there slow database queries?
- Are there external API calls causing delays?

**Screenshot**: Save as `images/06-performance-bottlenecks.png`

### 4.4 Examine Service Dependencies

**Steps**:
1. Click on **Dependencies** tab
2. Observe service dependency graph
3. Click on connection lines to see:
   - Call volume between services
   - Average latency
   - Error rate

## 🔔 Section 7: Configure Alerts

## 📊 Key Metrics to Monitor

### Microservices Metrics
- **Request Rate**: Requests per second for each service
- **Latency**: P50, P95, P99 response times
- **Error Rate**: Percentage of failed requests
- **Saturation**: CPU, memory, and network utilization

### AI Agent Metrics
- **Invocation Rate**: How often AI agent is triggered
- **Processing Time**: Time to analyze and respond
- **API Call Latency**: watsonx.ai response times
- **Decision Accuracy**: Correct vs incorrect predictions

### Database Metrics
- **Query Performance**: Slow query detection
- **Connection Pool**: Active vs idle connections
- **Transaction Rate**: Commits and rollbacks
- **Lock Contention**: Blocking queries
