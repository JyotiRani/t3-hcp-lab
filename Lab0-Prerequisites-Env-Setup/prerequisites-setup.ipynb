## 1️⃣ Prerequisites Installation

### 1.1 : Verify Tool Installations

```python
# Verify Terraform
print("Terraform Version:")
!terraform version 2>/dev/null || echo "❌ Terraform not installed"

print("\n" + "="*90)
```

```python
# Verify Ansible
print("Ansible Version:")
!ansible --version 2>/dev/null || echo "❌ Ansible not installed"

print("\n" + "="*90)
```

```python
# Verify kubectl
print("kubectl Version:")
!kubectl version --client 2>/dev/null || echo "❌ kubectl not installed"

print("\n" + "="*90)
```

```python
# Verify Git
print("Git Version:")
!git --version

print("\n" + "="*90)
```

```python
# Verify IBM Cloud CLI
print("IBM Cloud CLI Version:")
!ibmcloud version 2>/dev/null || echo "❌ IBM Cloud CLI not installed"

print("\n" + "="*90)
```

### 1.2 : Ansible Connectivity Test

```python
# Test Ansible connectivity
print("Testing Ansible connectivity...\n")
!ansible localhost -m ping

print("\n✅ Expected output: localhost | SUCCESS => { 'ping': 'pong' }")

print("\n" + "="*90)
```
